```python
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]  # utilizamos 2 features para desplegar los datos en 2D
y = iris.target

def crear_mesh(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy

def desplegar_mesh(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

## Kernel Lineal

```python
model = svm.SVC(kernel='linear')
clf = model.fit(X, y)

fig, ax = plt.subplots()
X0, X1 = X[:, 0], X[:, 1]
xx, yy = crear_mesh(X0, X1)

desplegar_mesh(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)

ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')

# visualizar los vectores de sorpote usados por el modelo
#ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80, facecolors='none', zorder=10, edgecolors='k')

plt.show()
```

## Kernel Polinomial

```python
model = svm.SVC(kernel='poly', degree=3)
clf = model.fit(X, y)

fig, ax = plt.subplots()
X0, X1 = X[:, 0], X[:, 1]
xx, yy = crear_mesh(X0, X1)

desplegar_mesh(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)

ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')

plt.show()
```

## Kernel Radial

```python
model = svm.SVC(kernel='rbf', gamma='auto') # intentar con gamma = 2
clf = model.fit(X, y)

fig, ax = plt.subplots()
X0, X1 = X[:, 0], X[:, 1]
xx, yy = crear_mesh(X0, X1)

desplegar_mesh(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)

ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
plt.show()
```
